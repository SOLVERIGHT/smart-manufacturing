```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.getOrCreate()

base_path = "/Workspace/Users/seth.ms07g@nctu.edu.tw/.bundle/oee-analytics/dev/files/data/sample"

quality_df = (
    spark.read
    .option("header", True)
    .option("inferSchema", True)
    .csv(f"{base_path}/quality_logs.csv")
)

downtime_df = (
    spark.read
    .option("header", True)
    .option("inferSchema", True)
    .csv(f"{base_path}/downtime_logs.csv")
)

print(f"Quality rows: {quality_df.count()}")
print(f"Downtime rows: {downtime_df.count()}")
```
